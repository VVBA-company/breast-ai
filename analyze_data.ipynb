LumA (Luminal A):
ƒê·∫∑c ƒëi·ªÉm: ƒê√¢y l√† lo·∫°i "l√†nh" nh·∫•t trong c√°c lo·∫°i ung th∆∞ v√∫. T·∫ø b√†o ph√°t tri·ªÉn ch·∫≠m, c√≥ th·ª• th·ªÉ n·ªôi ti·∫øt (ER d∆∞∆°ng t√≠nh).
ƒêi·ªÅu tr·ªã: ƒê√°p ·ª©ng r·∫•t t·ªët v·ªõi li·ªáu ph√°p n·ªôi ti·∫øt, ti√™n l∆∞·ª£ng s·ªëng t·ªët nh·∫•t.

LumB (Luminal B):

ƒê·∫∑c ƒëi·ªÉm: C≈©ng c√≥ th·ª• th·ªÉ n·ªôi ti·∫øt nh∆∞ LumA nh∆∞ng "d·ªØ" h∆°n. T·∫ø b√†o ph√¢n chia nhanh h∆°n (ch·ªâ s·ªë Ki67 cao) ho·∫∑c c√≥ th·ªÉ k√®m theo HER2 d∆∞∆°ng t√≠nh.
Ti√™n l∆∞·ª£ng: K√©m h∆°n LumA, th∆∞·ªùng c·∫ßn ƒëi·ªÅu tr·ªã m·∫°nh tay h∆°n.

Her2 (HER2-enriched):

ƒê·∫∑c ƒëi·ªÉm: Nh√≥m n√†y t·∫ø b√†o c√≥ s·ª± b√πng n·ªï c·ªßa gen HER2. Kh·ªëi u ph√°t tri·ªÉn r·∫•t nhanh v√† m·∫°nh.
L∆∞u √Ω: V·ªÅ m·∫∑t sinh h·ªçc ph√¢n t·ª≠, nh√≥m n√†y kh√°c v·ªõi x√©t nghi·ªám HER2 l√¢m s√†ng (IHC/FISH), nh∆∞ng ƒë·∫∑c t√≠nh chung l√† √°c t√≠nh cao.

claudin-low:

ƒê·∫∑c ƒëi·ªÉm: M·ªôt ph√¢n nh√≥m kh√° ƒë·∫∑c bi·ªát v√† hi·∫øm. N√≥ th∆∞·ªùng c√≥ ƒë·∫∑c t√≠nh gi·ªëng t·∫ø b√†o g·ªëc (stem cell), th∆∞·ªùng thu·ªôc nh√≥m "B·ªô ba √¢m t√≠nh" (Triple Negative).
Ti√™n l∆∞·ª£ng: Kh√≥ ƒëi·ªÅu tr·ªã, ƒë√°p ·ª©ng k√©m v·ªõi h√≥a tr·ªã chu·∫©n.

Basal (Basal-like):

ƒê·∫∑c ƒëi·ªÉm: K·∫ª th√π nguy hi·ªÉm nh·∫•t. ƒêa s·ªë tr√πng v·ªõi nh√≥m "B·ªô ba √¢m t√≠nh" (ER-, PR-, HER2-). Hay g·∫∑p ·ªü ng∆∞·ªùi tr·∫ª, ƒë·ªôt bi·∫øn gen BRCA1.
Ti√™n l∆∞·ª£ng: X·∫•u nh·∫•t, ph√°t tri·ªÉn c·ª±c nhanh, ch·ªâ c√≥ th·ªÉ d√πng h√≥a tr·ªã (do kh√¥ng c√≥ th·ª• th·ªÉ ƒë·ªÉ d√πng thu·ªëc ƒë√≠ch hay n·ªôi ti·∫øt).

Normal (Normal-like):

ƒê·∫∑c ƒëi·ªÉm: H·ªì s∆° gen nh√¨n kh√° gi·ªëng m√¥ v√∫ b√¨nh th∆∞·ªùng (m√¥ m·ª° v√† m√¥ li√™n k·∫øt). Th∆∞·ªùng c√≥ ti√™n l∆∞·ª£ng trung b√¨nh-t·ªët.
NC (Not Classified): Kh√¥ng ph√¢n lo·∫°i ƒë∆∞·ª£c (l·ªói d·ªØ li·ªáu).

PAM50 l√† m·ªôt x√©t nghi·ªám gen hi·ªán ƒë·∫°i, ph√¢n t√≠ch m·ª©c ƒë·ªô ho·∫°t ƒë·ªông c·ªßa 50 lo·∫°i gen c·ª• th·ªÉ trong t·∫ø b√†o kh·ªëi u. D·ª±a v√†o ƒë√≥, n√≥ chia ung th∆∞ v√∫ th√†nh c√°c nh√≥m c√≥ ƒë·∫∑c ƒëi·ªÉm sinh h·ªçc v√† c√°ch ƒëi·ªÅu tr·ªã kh√°c nhau. ƒê√¢y l√† c√°ch ph√¢n lo·∫°i s√¢u h∆°n v√† ch√≠nh x√°c h∆°n v·ªÅ m·∫∑t sinh h·ªçc so v·ªõi c√°c ch·ªâ s·ªë l√¢m s√†ng th√¥ng th∆∞·ªùng (nh∆∞ ER, HER2 ƒë∆°n l·∫ª).

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# --- PH·∫¶N 1: X·ª¨ L√ù D·ªÆ LI·ªÜU (Code c·ªßa b·∫°n) ---
mutation_cols = [c for c in df.columns if c.endswith("_mut")]
clinical_outcome = df["overall_survival_months"]

# Chuy·ªÉn ƒë·ªïi d·ªØ li·ªáu ƒë·ªôt bi·∫øn sang 0 v√† 1
for col in mutation_cols:
    # N·∫øu l√† '0' th√¨ l√† 0, n·∫øu c√≥ gi√° tr·ªã kh√°c (v√† kh√¥ng ph·∫£i NaN) th√¨ l√† 1
    df[col] = df[col].apply(lambda x: 0 if str(x).strip() == '0' else (1 if pd.notna(x) else pd.NA)).astype('Int64')

# T√≠nh to√°n t∆∞∆°ng quan
corr_list = []
for col in mutation_cols:
    # L∆∞u √Ω: df[col] c·∫ßn chuy·ªÉn v·ªÅ float ƒë·ªÉ t√≠nh corr, dropna ƒë·ªÉ lo·∫°i b·ªè gi√° tr·ªã thi·∫øu
    if df[col].notna().sum() > 0: # Ch·ªâ t√≠nh n·∫øu c·ªôt c√≥ d·ªØ li·ªáu
        corr = abs(df[col].astype(float).corr(clinical_outcome))
        corr_list.append((col, corr))

corr_df = pd.DataFrame(corr_list, columns=["mutation", "corr_with_survival"])
# S·∫Øp x·∫øp gi·∫£m d·∫ßn v√† l·∫•y Top 15
corr_df = corr_df.sort_values(by="corr_with_survival", ascending=False).head(15)

# --- PH·∫¶N 2: V·∫º BI·ªÇU ƒê·ªí ---
plt.figure(figsize=(10, 8)) # Thi·∫øt l·∫≠p k√≠ch th∆∞·ªõc ·∫£nh

# V·∫Ω bi·ªÉu ƒë·ªì thanh ngang
# Palette 'magma_r' ho·∫∑c 'viridis' gi√∫p nh·∫•n m·∫°nh c√°c thanh c√≥ gi√° tr·ªã cao
ax = sns.barplot(data=corr_df, y='mutation', x='corr_with_survival', palette='viridis')

# Trang tr√≠ bi·ªÉu ƒë·ªì
plt.title('Top 15 ƒê·ªôt bi·∫øn c√≥ t∆∞∆°ng quan m·∫°nh nh·∫•t v·ªõi Th·ªùi gian s·ªëng c√≤n', fontsize=16)
plt.xlabel('H·ªá s·ªë t∆∞∆°ng quan tuy·ªát ƒë·ªëi (|Correlation|)', fontsize=12)
plt.ylabel('T√™n ƒë·ªôt bi·∫øn', fontsize=12)

# Th√™m gi√° tr·ªã s·ªë l√™n ƒë·∫ßu m·ªói thanh
for container in ax.containers:
    ax.bar_label(container, fmt='%.3f', padding=3)

plt.grid(axis='x', linestyle='--', alpha=0.7) # Th√™m l∆∞·ªõi m·ªù tr·ª•c x cho d·ªÖ nh√¨n
plt.tight_layout() # T·ª± ƒë·ªông cƒÉn ch·ªânh l·ªÅ
plt.show()
```

Tuy nhi√™n, v·∫´n t·ªìn t·∫°i m·ªôt nh√≥m ngo·∫°i lai ƒë√°ng k·ªÉ c√≥ s·ªë l∆∞·ª£ng h·∫°ch di cƒÉn r·∫•t cao nh∆∞ng kh√¥ng h√≥a tr·ªã, ƒëi·ªÅu n√†y g·ª£i √Ω r·∫±ng quy·∫øt ƒë·ªãnh ƒëi·ªÅu tr·ªã c√≤n ph·ª• thu·ªôc v√†o c√°c y·∫øu t·ªë kh√°c nh∆∞ tu·ªïi t√°c, th·ªÉ tr·∫°ng b·ªánh nh√¢n ho·∫∑c ph√¢n nh√≥m sinh h·ªçc c·ªßa kh·ªëi u ch·ª© kh√¥ng ch·ªâ d·ª±a v√†o s·ªë l∆∞·ª£ng h·∫°ch.

```python


# L·∫•y danh s√°ch c√°c c·ªôt ƒë·ªôt bi·∫øn gene
mutation_cols = [col for col in df.columns if col.endswith('_mut')]

# T√≠nh t·ªïng s·ªë l∆∞·ª£ng ƒë·ªôt bi·∫øn cho m·ªói gene
mutation_counts = df[mutation_cols].sum().sort_values(ascending=False)

# V·∫Ω bi·ªÉu ƒë·ªì top N ƒë·ªôt bi·∫øn ph·ªï bi·∫øn nh·∫•t
N = 20  # S·ªë l∆∞·ª£ng ƒë·ªôt bi·∫øn hi·ªÉn th·ªã
top_mutations = mutation_counts.head(N)

plt.figure(figsize=(12, 6))
sns.barplot(x=top_mutations.values, y=top_mutations.index)
plt.title(f'Top {N} ƒê·ªôt bi·∫øn Gene Ph·ªï bi·∫øn nh·∫•t')
plt.xlabel('S·ªë l∆∞·ª£ng')
plt.ylabel('Gene')
plt.show()

# T·∫°o ma tr·∫≠n t∆∞∆°ng quan gi·ªØa c√°c ƒë·ªôt bi·∫øn (ch·ªâ hi·ªÉn th·ªã top 10)
top_mutation_cols = top_mutations.head(10).index
correlation_matrix = df[top_mutation_cols].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('T∆∞∆°ng quan gi·ªØa C√°c ƒê·ªôt bi·∫øn Gene (Top 10)')
plt.show()
```
